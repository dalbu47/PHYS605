```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

filelist = os.listdir(path='CSV_Data')

#print(filelist)
data = []
for i in range(len(filelist)):
    file_path = f'CSV_Data/{filelist[i]}'
    read = pd.read_csv(file_path, comment='#', encoding='latin1')
    data.append(read)
    print(f"{filelist[i]}:")
    display(data[i].head())

```

```python
for i in [0, 1, 6, 8]:
    filelist[i] = 'Magnitude'
for i in [7,9]:
    filelist[i] = 'Magnitude (Stable)'
for i in [2,3]:
    filelist[i] = 'Triangle Wave (1kHz, 100mV)'
for i in [4,5]:
    filelist[i] = 'Sine Wave (100kHz, 20mV)'
for i in [10,11]:
    filelist[i] = 'Triangle Wave (100kHz, 20mV)'
for i in range(len(data)):
    file_name = filelist[i]
    plt.figure(figsize=(12, 6))

    num_columns = len(data[i].columns)

    if num_columns == 3: # for 3 column data files
        if data[i].columns[2] == 'Channel 2 Phase (째)':
            # Plot bode and phase separately from voltage
            plt.figure(figsize=(10, 6))
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 1], label=data[i].columns[1])
            plt.xscale('log')
            plt.title(f'Plot for {file_name}')
            plt.xlabel(data[i].columns[0])
            plt.ylabel(data[i].columns[1])
            plt.legend()
            plt.figure(figsize=(10,6))
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 2], color='g', label=data[i].columns[2])
            plt.title(f'Phase Plot for {file_name}')
            plt.xlabel(data[i].columns[0])
            plt.ylabel("Phase (deg)")
            plt.legend()
        else:
            # Plot for Voltage
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 1], label=data[i].columns[1])
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 2], label=data[i].columns[2])
            plt.title(f'Voltage Plot for {file_name}')
            plt.xlabel(data[i].columns[0])
            plt.ylabel("Voltage (V)")
            plt.legend()

    elif num_columns == 4: # for 4 column data files
            # Voltage and Amp Plot
        if data[i].columns[3] == 'Math 1 (V)':
            plt.figure(figsize=(10, 6))
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 1], label=data[i].columns[1])
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 3], label=data[i].columns[3])
            plt.xscale('log')
            plt.title(f'Voltage Plot for {file_name}')
            plt.xlabel(data[i].columns[0])
            plt.ylabel(data[i].columns[2])
            plt.legend()
               
        else:
            # Bode and Gain + Phase Plot
            plt.figure(figsize=(10, 6))
            #plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 1], label=data[i].columns[1])
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 2], label=data[i].columns[2])
            plt.xscale('log')
            plt.title(f'Plot for {file_name}')
            plt.xlabel(data[i].columns[0])
            plt.ylabel(data[i].columns[2])
            plt.legend()
        
            plt.figure(figsize=(10, 6))
            plt.plot(data[i].iloc[:, 0], data[i].iloc[:, 3], color='g', label=data[i].columns[3])
            plt.xscale('log')
            plt.title(f'Phase Plot for {file_name}')
            plt.xlabel(data[i].columns[0])
            plt.ylabel("Phase (deg)")
            plt.legend()

        plt.tight_layout()

    plt.show()
```
