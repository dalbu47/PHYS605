# $\text{Plotting Program for Logic Binary Counting Sequence}$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file = '8bitOutput(Logic Analyzer).csv'  # idk why i put this here
data = []  # empty list to store cs file data as an array (pandas does all the heavy lifting)

read = pd.read_csv(file, comment='#', encoding='latin1')  # reads file and ignores the stuff starting with # in said file
data.append(read)  # appends data to the list
display(data[0].head)  # displays some of the array just to show what we got
```

```python
plt.figure(figsize=(15,8))  # appropriately sized figure
for i in range(8):  # for loop goes through each column of the array and plots it on one figure
    plt.plot(data[0].iloc[:, 0], data[0].iloc[:, i+1]-(1.1*i), label=data[0].columns[i+1])  # I shift each sequential column down 
plt.title('Logic Binary Counting Sequence')
plt.legend()  # creates legend
plt.xticks([])  # gets rid of x ticks and numbers
plt.yticks([])  # gets rid of y ticks and numbers
plt.xlim(data[0].iloc[100, 0], data[0].iloc[3200, 0])  # limits x-axis to make plot look better and less cramped
```
