```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os


filelist = os.listdir(path='CSV_Data')

data = []

file_path = f'CSV_Data/{filelist[0]}'
read = pd.read_csv(file_path, comment='#', encoding='latin1')
data.append(read)
print(f"{filelist[0]}:")
display(data[0].head())
```

```python
plt.figure(figsize=(10, 6)) # I-V Curve
plt.plot(data[0].iloc[:, 0], data[0].iloc[:, 1], label=data[0].columns[1], color='cyan')
plt.title(f'Current vs Voltage', fontsize='large')
plt.xlabel('Voltage(V)')
plt.ylabel('Current(A)')
plt.legend()

# Had to delete first bit until we got to ~2.5 then deleted the bit after that went to ~-0.75
# this was due to some endpoints connecting in the data
```
