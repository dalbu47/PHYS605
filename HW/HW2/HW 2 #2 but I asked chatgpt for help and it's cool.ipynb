```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

files = ['1N4148 diode IV data HW 2.csv', 'Zener diode IV data HW 2.csv', '1N5223B 3.3V diode IV data HW 2.csv', 'BZT52C2V7  diode IV data HW 2.csv']
namelist = ['1N4148', 'Zener', '1N5223B 3.3V', '1BZT52C2V7']

fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, 1, figsize=(8, 15))
fig.subplots_adjust(hspace=0.7)
exdatalist = []

# Read each CSV file and plot individual I-V curves
colorlist = ['r', 'c', 'g', 'm']
for i, file_name in enumerate(files):
    exdata = pd.read_csv(file_name)
    exdatalist.append(exdata)

    ax = [ax1, ax2, ax3, ax4][i]
    ax.plot(exdata['Voltage(V)'], exdata['Current(A)'], label=f'{namelist[i]}', color=colorlist[i])
    ax.set_xlabel('Voltage', fontsize=12)
    ax.set_ylabel('Current', fontsize=12)
    ax.legend(fontsize=7)
    ax.set_title(f"I-V curve for {namelist[i]} Diode", fontsize=15)

# Plot combined I-V curve
combined_currents = [exdata['Current(A)'].tolist() for exdata in exdatalist]
for j in range(len(files)):
    ax5.plot(exdatalist[j]['Voltage(V)'], combined_currents[j], label=f'Combined - {namelist[j]}', color=colorlist[j])

ax5.set_xlabel('Voltage', fontsize=12)
ax5.set_ylabel('Current', fontsize=12)
ax5.legend(fontsize=6, loc='upper left')  # Set the legend position to upper right
ax5.set_title("I-V curve for all Diodes", fontsize=15)

plt.show()
```
