```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

files = ['1N4148 diode IV data HW 2.csv', 'Zener diode IV data HW 2.csv', '1N5223B 3.3V diode IV data HW 2.csv', 'BZT52C2V7  diode IV data HW 2.csv']
namelist = ['1N4148', 'Zener', '1N5223B 3.3V', '1BZT52C2V7']

fig, (ax1,ax2,ax3, ax4)=plt.subplots(4,1, figsize=(8,10))
fig.subplots_adjust(hspace=0.7)
exdatalist = []
for i in range(len(files)):
    exdata = pd.read_csv(f'{files[i]}')
    exdatalist.append(exdata)
    ax = [ax1,ax2,ax3, ax4][i]
    colorlist = ['r', 'c', 'g', 'm']
    ax.plot(exdatalist[i]['Voltage(V)'], exdatalist[i]['Current(A)'], label=f'{namelist[i]}', color=colorlist[i])
    ax.set_xlabel('Voltage', fontsize=12)
    ax.set_ylabel('Current', fontsize=12)
    ax.legend(fontsize=7)
    ax.set_title(f"I-V curve for {namelist[i]} Diode", fontsize=15)


print(f"The curve for the {namelist[0]} diode is much steeper than any of the others, also starting at -10 V and rising much higher than any of the other curves. This shows an exponential increase in current as voltage increases linearly. The curve for the {namelist[1]} diode shows an initial exponential decay of current as voltage increases but a plateau for about 4V until it starts increasing again. This could have some useful properties, perhaps for keeping current stable over a change in voltage? The curve for the {namelist[2]} diode shows A similar curve to the {namelist[1]} diode but has a more linear 'ramp' up to the plateau zone and a linear ramp up after the plateau. It's range is also a little shorter than {namelist[1]} diode. The curve for the {namelist[3]} diode has the second longest range after the {namelist[0]} diode, and the slope leading up to the plateau is less steep than the {namelist[2]} diode. The {namelist[2]} and {namelist[3]} diodes are both {namelist[1]} diodes, but the way that they reach the current plateau is different.")

print("I played around with the simulation (with the zener diode) a bit, and I discovered some things:")

my_list = [
    "Voltage source",
    "   The Ramp high and low values for the ramp setting will change the voltage and current plots' slope, higher values increase the slope and lower values decrease the slope" ,
    "   Changing the ramp time will do some strange things. Its at 10 by default but if you set it to 5 both plots become symmetric over the y axis? I tried making it 1 and it rose quickly to a peak then slowly decreased. I can't quite seem to figure out what it means",
    "   I discovered the ramp period must be \u2265 than the ramp time, which makes some sense since period relies on the ramp arriving at the peak before the ramp restarts",
    "   Different functions for the Voltage source yielded interesting plots, with the sine one giving sinusoidal periodic waves for voltage and plateau-esque periodic waves for current. Pulse gave me square waves for voltage and for current it gave me a cool looking function that I decided to plot, which I will submit with the file name: 'Zener Square Wave Vs Current Graph.pdf'."
]

formatted_list = '\n'.join(f'• {item}' if not item.startswith('   ') else f'   • {item.strip()}' for item in my_list)

print(formatted_list)

#fig = plt.figure(figsize=(100,100))
#data = pd.read_csv('Zener diode IT data  square wave.csv')
#plt.plot(data['Time(s)'], data['Current(A)'])
```
